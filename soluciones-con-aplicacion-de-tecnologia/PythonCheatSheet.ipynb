## 3. Números y operaciones

Al igual que con otros lenguajes de programación, hay cuatro tipos de números: 
- Enteros (por ejemplo, 1, 20, 45, 1000) indicados por *ent*
- Números de coma flotante (por ejemplo, 1,25; 20,35; 1000,00) indicados por *flotante*
- Enteros largos 
- Números complejos (por ejemplo, x+2y donde se conoce x)

Operación | Resultado
----------------|-------------------------------------               
x + y | Suma de x e y	
x - y | Diferencia de x e y	
x * y | Producto de x e y	
x / y | Cociente de x e y
x // y | Cociente de x e y (piso)
x % y | Resto de x / y
abs(x) | Valor absoluto de x	
ent(x) | x convertido a entero
largo(x) | x convertido a entero largo
flotante(x) | x convertido a coma flotante	
pot(x, y) | x a la potencia y	
x ** y | x a la potencia y	

```python
# Number examples
a = 5 + 8
print("Sum of int numbers: {} and number format is {}".format(a, type(a)))

b = 5 + 2.3
print ("Sum of int and {} and number format is {}".format(b, type(b)))
```

## 4. Manipulación de cadenas

Python tiene características ricas como otros lenguajes de programación para la manipulación de cadenas.

```python
# Store strings in a variable
test_word = "hello world to everyone"

# Print the test_word value
print(test_word)

# Use [] to access the character of the string. The first character is indicated by '0'.
print(test_word[0])

# Use the len() function to find the length of the string
print(len(test_word))

# Some examples of finding in strings
print(test_word.count('l')) # Count number of times l repeats in the string
print(test_word.find("o")) # Find letter 'o' in the string. Returns the position of first match.
print(test_word.count(' ')) # Count number of spaces in the string
print(test_word.upper()) # Change the string to uppercase
print(test_word.lower()) # Change the string to lowercase
print(test_word.replace("everyone","you")) # Replace word "everyone" with "you"
print(test_word.title()) # Change string to title format
print(test_word + "!!!") # Concatenate strings
print(":".join(test_word)) # Add ":" between each character
print("".join(reversed(test_word))) # Reverse the string 
```

## 5. Listas, tuplas y diccionarios

Python admite listas de tipos de datos, tuplas, diccionarios y matrices.

### Listas

Se crea una lista colocando todos los elementos (artículos) entre corchetes \[ ] separados por comas. Una lista puede tener cualquier cantidad de elementos, y pueden ser de diferentes tipos (entero, flotante, cadenas, etc.).

```python
# A Python list is similar to an array. You can create an empty list too.

my_list = []

first_list = [3, 5, 7, 10]
second_list = [1, 'python', 3]
```

```python
# Nest multiple lists
nested_list = [first_list, second_list]
nested_list
```

```python
# Combine multiple lists
combined_list = first_list + second_list
combined_list
```

```python
# You can slice a list, just like strings
combined_list[0:3]
```

```python
# Append a new entry to the list
combined_list.append(600)
combined_list
```

```python
# Remove the last entry from the list
combined_list.pop()
```

```python
# Iterate the list
for item in combined_list:
    print(item)    
```

### Tuplas

Una tupla es similar a una lista, pero las usa con paréntesis ( ) en lugar de corchetes. La principal diferencia es que una tupla es inmutable, mientras que una lista es mutable.

```python
my_tuple = (1, 2, 3, 4, 5)
my_tuple[1:4]
```

### Diccionarios

Un diccionario también se conoce como matriz asociativa. Un diccionario consiste en una colección de pares clave-valor. Cada par clave-valor asigna la clave a su valor asociado.

```python
desk_location = {'jack': 123, 'joe': 234, 'hary': 543}
desk_location['jack']
```

## 7. Bucles
**Bucle si, entonces, eLiF**: Python admite declaraciones condicionales como cualquier otro lenguaje de programación. Python se basa en la sangría (espacio en blanco al principio de la línea) para definir el alcance del código. 

```python
a = 22
b = 33
c = 100

# if ... else example
if a > b:
    print("a is greater than b")
else:
    print("b is greater than a")
    
    
# if .. else .. elif example

if a > b:
    print("a is greater than b")
elif b > c:
    print("b is greater than c")
else:
    print("b is greater than a and c is greater than b")
```

**Bucle mientras:** ejecuta un conjunto de declaraciones siempre y cuando la condición sea verdadera

```python
# Sample while example
i = 1
while i < 10:
    print("count is " + str(i))
    i += 1

print("="*10)

# Continue to next iteration if x is 2. Finally, print message once the condition is false.

x = 0
while x < 5:
    x += 1
    if x == 2:
        continue
    print(x)
else:
    print("x is no longer less than 5")
```

**Bucle para:** Un bucle `For` es más como un iterador en Python. Un bucle `For` se utiliza para iterar sobre una secuencia (lista, tupla, diccionario, conjunto, cadena o rango).

```python
# Sample for loop examples
fruits = ["orange", "banana", "apple", "grape", "cherry"]
for fruit in fruits:
    print(fruit)

print("\n")
print("="*10)
print("\n")

# Iterating range
for x in range(1, 10, 2):
    print(x)
else:
    print("task complete")

print("\n")
print("="*10)
print("\n")

# Iterating multiple lists
traffic_lights = ["red", "yellow", "green"]
action = ["stop", "slow down", "go"]

for light in traffic_lights:
    for task in action:
        print(light, task)
```

## 8. Administración de archivos
La función clave para trabajar con archivos en Python es la función `open()`. La función `open()` toma dos parámetros: nombre de archivo y modo.

Existen cuatro métodos (modos) diferentes para abrir un archivo:

- "r": Leer
- "a": Agregar
- "w": Escribir
- "x": Crear

Además, puede especificar si el archivo debe ser manejado en modo binario o texto.

- "t": Texto
- "b": Binario

```python
# Let's create a test text file
!echo "This is a test file with text in it. This is the first line." > test.txt
!echo "This is the second line." >> test.txt
!echo "This is the third line." >> test.txt
```

```python
# Read file
file = open('test.txt', 'r')
print(file.read())
file.close()

print("\n")
print("="*10)
print("\n")

# Read first 10 characters of the file
file = open('test.txt', 'r')
print(file.read(10))
file.close()

print("\n")
print("="*10)
print("\n")

# Read line from the file

file = open('test.txt', 'r')
print(file.readline())
file.close()
```

```python
# Create new file

file = open('test2.txt', 'w')
file.write("This is content in the new test2 file.")
file.close()

# Read the content of the new file
file = open('test2.txt', 'r')
print(file.read())
file.close()
```

```python
# Update file
file = open('test2.txt', 'a')
file.write("\nThis is additional content in the new file.")
file.close()

# Read the content of the new file
file = open('test2.txt', 'r')
print(file.read())
file.close()
```

```python
# Delete file
import os
file_names = ["test.txt", "test2.txt"]
for item in file_names:
    if os.path.exists(item):
        os.remove(item)
        print(f"File {item} removed successfully!")
    else:
        print(f"{item} file does not exist.")
```

## 9. Funciones

Una función es un bloque de código que se ejecuta cuando se llama. Puede pasar datos, o *parámetros*, a la función. En Python, una función está definida por `def`.

```python
# Defining a function
def new_funct():
    print("A simple function")

# Calling the function
new_funct()
```

```python
# Sample fuction with parameters

def param_funct(first_name):
    print(f"Employee name is {first_name}.")

param_funct("Harry")
param_funct("Larry")
param_funct("Shally")
```

**Funciones anónimas (lambda):** Una lambda es una pequeña función anónima. Una función lambda puede tomar cualquier número de argumentos, pero solo una expresión.

```python
# Sample lambda example
x = lambda y: y + 100
print(x(15))

print("\n")
print("="*10)
print("\n")

x = lambda a, b: a*b/100
print(x(2,4))
```

## 10. Trabajo con fecha y hora 

Se puede usar un módulo `datetime` en Python para trabajar con objetos de fecha.

```python
import datetime

x = datetime.datetime.now()

print(x)
print(x.year)
print(x.strftime("%A"))
print(x.strftime("%B"))
print(x.strftime("%d"))
print(x.strftime("%H:%M:%S %p"))
```
