5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:
*   Numéricas: `Age, Number of siblings`
*   Categóricas: `Gender, Education`

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas. NO olvides incluir las conclusiones del análisis con los resultados obtenidos.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'CON valores faltantes',False:'SIN valores faltantes'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])

    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(10,4))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'CON valores faltantes',False:'SIN valores faltantes'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
numerical_attributes = ['Age', 'Number of siblings']
BM_MV = df.Weight.isna()

for attribute in numerical_attributes:
    print('Analisis de los valores faltantes en {}:'.format(attribute))
    Diagnose_MV_Numerical(df, attribute, BM_MV)
    print('----------------------------------------------------------')
```

7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
plt.boxplot(df.Weight.dropna())
plt.show()
```

```python
Q1 = df.Weight.quantile(0.25)
Q3 = df.Weight.quantile(0.75)
IQR = Q3-Q1
BM = (df.Weight > (Q3+1.5 *IQR)) | (df.Weight < (Q1-1.5 *IQR))
weight_outliers_df = df[BM]
len(weight_outliers_df)
```

Existen 10 valores atipicos en la columna peso. Esto lo confirmamos viendo la grafica boxplot, en donde todos aquellos circulos (fliers) afuera de los "bigotes" representan los valores atipicos.

Por otro lado, para verificar el numero exacto de valores atipicos, podemos utilizar el metodo IQR, en donde usamos una mascara booleana para obtener todos aquellos valores de la columna peso que sean menores a Q1-1.5*IQR y mayores a Q3+1.5 *IQR.

Por ultimo estoy almacenando esos registros en un nuevo DataFrame llamado "weight_outliers_df".

Y ahora aplicare una mascara booleana invertida para quitar los valores atipicos del DataFrame original.

```python
# Eliminar los registros filtrados del DataFrame original
df = df[~BM]
len(df.Weight)
```

9. ¿Hay encuestados que sean atípicos según su índice de masa corporal (`BMI`)? El `BMI` puede ser calculado usando la siguiente fórmula:

`BMI = Weight / Height ^ 2`

**NOTA**. El peso tiene que estar en kilogramos y la altura en metros, pero en el conjunto la altura se ha registrado en centímetros, por lo que deberás convertirla a metros.

```python
# Primero verificar los valores faltantes en la altura
df.Height.isna().sum()
```

Encontramos que faltan 19 valores en la columna de la altura. Asi que habra que aplicar una imputacion usando la media, ya que por naturaleza la altura tambien tiene una distribucion normal (gaussiana). Asi los valores faltantes seran reemplazados por el promedio general de la altura, sin afectar el sesgo de la muestra.

```python
df.Height.fillna(value=df.Height.mean(), inplace=True)
df.Height.isna().sum()
```

```python
# Ahora si podemos convertir la altura de cm a metros
df.Height = df.Height / 100

# Crear una columna que almacene el índice de masa corporal (BMI)
df['BMI'] = df.Weight / (df.Height ** 2)
```

```python
# Verificar que los calculos sean correctos
print(df[['Weight', 'Height', 'BMI']].head(10))
```

```python
df[['Weight', 'Height', 'BMI']].describe().T
```

```python
# Verificar valores atipicos con el metodo IQR
Q1_I = df.BMI.quantile(0.25)
Q3_I = df.BMI.quantile(0.75)
IQR_I = Q3_I-Q1_I
BM_I = (df.BMI > (Q3_I+1.5 *IQR_I)) | (df.BMI < (Q1_I-1.5 *IQR_I))

len(df[BM_I])
```
